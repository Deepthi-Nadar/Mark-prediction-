**Linear Regression - Mark prediction**

Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load the Dataset

```python
dataset = pd.read_csv("/content/Rounded_Student_Hours_Studied_vs_Marks_Dataset.csv")
dataset.head()
```

Data Analysis

```python
dataset.describe()
```

```python
#Data Distribution

plt.scatter(dataset['Hours_Studied'], dataset['Marks'])
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Distribution of Hours Studied vs Marks")
plt.show()
```

Split Independent & Dependent Variables

```python
X = dataset[['Hours_Studied']]   # Independent variable
y = dataset['Marks']             # Dependent variable
```

Split into Training & Testing Sets (80% / 20%)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

Train the Simple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Predict the Results

```python
y_train_pred = regressor.predict(X_train)
y_test_pred = regressor.predict(X_test)
```

Plot Training Set Results

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, y_train_pred)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Training Set: Hours Studied vs Marks")
plt.show()
```

Plot Test Set Results

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Test Set: Hours Studied vs Marks")
plt.show()
```

Coefficient & Intercept

```python
print("Coefficient (Slope):", regressor.coef_[0])
print("Intercept:", regressor.intercept_)
```

Accuracy of the Model (R Score)

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_test_pred)
print("R Square Value:", r2 * 100)
```

Predict Marks for New Study Hours

```python
hours = [[7]]  # Example: 7 hours studied
predicted_marks = regressor.predict(hours)

print("Predicted Marks for 7 hours:", predicted_marks[0])
```
